```python
# Calculate a grid of points from -2-2 with 100 points in each direction
x,y = np.meshgrid(np.linspace(-2,2,100),np.linspace(-2,2,100));

# Calculate the scalar field
f = x**2 + y**2

# False colour plot
#cmap options are 'magma', 'plasma', 'flag', 'Spectral', 'autumn', 'winter'
plt.pcolor(x,y,f, shading='auto', cmap='magma')#default cmap is viridis
plt.colorbar()
plt.contour(x,y,f)
plt.show()
```

```python
# Contour plot
#limits = np.linspace(0, np.amax(f), 10)#create contour values
plt.contour(x,y,f)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Surface plot
ax = plt.figure().add_subplot(projection='3d')
ax.plot_surface(x,y,f)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

```python
# Calculate a grid of points from -2-2 with 11 points in each direction
[x,y] = np.meshgrid(np.linspace(-2,2,10),np.linspace(-2,2,10));

# Calculate the vector field (source field)
fx = x#/np.sqrt(x**2+y**2)
fy = y#/np.sqrt(x**2+y**2)

colors = np.abs(np.sqrt(fx**2+fy**2))#create colors based on magnitude of vectors

# Plot arrows
plt.quiver(x, y, fx, fy, colors, cmap='Spectral') #cmap options are 'magma', 'plasma', 'flag', 'Spectral', 'autumn', 'winter'
#plt.quiver(x, y, fx, fy)
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### 3D Vector Fields
Suppose we wanted to look at a three-dimensional vector field. Here is an example.


```python
# Calculate the vector field (source field)
gx = y
gy = 0

colors = np.abs(np.sqrt(gx**2+gy**2))#create colors based on magnitude of vectors

# Plot arrows
plt.quiver(x, y, gx, gy, colors, cmap='Spectral') #cmap options are 'magma', 'plasma', 'flag', 'Spectral', 'autumn', 'winter'
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Calculate a grid of points from -2-2 with 5 points in each direction
[x,y,z] = np.meshgrid(np.linspace(-2,2,6),np.linspace(-2,2,6),np.linspace(-2,2,6))

# Calculate the vector field (source field)

fx = x#/(x**2+y**2+z**2)**(3/2)
fy = y#/(x**2+y**2+z**2)**(3/2)
fz = z#/(x**2+y**2+z**2)**(3/2)

# Plot arrows
ax = plt.figure().add_subplot(projection='3d')
ax.quiver(x, y, z, fx, fy, fz, length=0.1, arrow_length_ratio=0.5, cmap='Spectral')  
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```
